```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()

axis = fig.add_axes([0.1,0.5,1,1]) #control the left, right , width, height of the canvas (from 0 to 1)

axis.plot(x,y, "r")
```

```python
# we can also draw a picture , or graph, inside another graph
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #big axes
axes2 = fig.add_axes([0.9,0.1,0.4,0.8]) #small canvas

axes1.plot(x,y, "r")
axes2.plot(y,x, "g")
```

```python


fig,axes = plt.subplots(dpi = 200)
axes.set_xlabel("x")
axes.set_ylabel("y")

axes.set_title("Title")

#axes.plot(x,y, "r")

axes.plot(x, x**2)
axes.plot(x, x**3)

axes.legend(["y = x**2", "y=x**3"],loc = 0)
```

```python
#In Matplotlib, you can set other properties such as line color, transparency, and more

fig, axes = plt.subplots(dpi=150)
axes.plot(x, x+1, color = "red",alpha=.8) #alpha is for less visible
axes.plot(x, x+2, color = "pink", linewidth = 3) #linewidth is the width of the line
axes.plot(x, x+3)
```

```python
fig, ax = plt.subplots(dpi=150)
ax.plot(x, x+1, color = "red",lw = 3, linestyle = "-") #lw is also for line width
ax.plot(x, x+2, color = "pink", lw = 3, linestyle = "-.") #linestyle is for the style of the line
ax.plot(x, x+3 , color="black",lw=3, linestyle = ":")

line, = ax.plot(x, x+4, color="orange", lw =1.5)
line.set_dashes([5,10,15,10])
```

# Other 2D Graphics

```python
n = np.arange(10)
fig, axes = plt.subplots(1,4,figsize = (16,5))
axes[0].set_title("Scatter Plot")
axes[0].scatter(n,n + 0.25*np.random.randn(len(n)))
axes[0].set_ylim([0,10])
axes[0].set_xlim([0,10])
axes[1].set_title("step")
axes[1].step(n, n *2, lw = 2)
axes[2].set_title("bar")
axes[2].bar(n,n**2,color = "red", align= "center", width = 0.4,alpha= 0.5)

axes[3].set_title("fill_between")
axes[3].fill_between(x,x**2,x**3,color= "green", alpha = 1)
```

```python
fig= plt.figure(figsize= (6,6))
ax = fig.add_axes([0.0,0.0,1,1], polar = True)
t = np.linspace(0,2 *np.pi, 100)
ax.plot(t,t**0.1, color = "blue", lw = 3)
```

```python
"""Histogram"""
n = np.random.randn(10000)
fig, axes = plt.subplots(1,2,figsize = (12,4))
axes[0].set_title("Histogram")
axes[0].hist(n)

axes[1].set_title("Cumulative detailed")
axes[1].hist(n, cumulative = True, bins=10)
```

# Draw contour image

```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

delta = 0.025
x = np.arange(-3.0,3.0,delta)
y = np.arange(-2.0,2.0, delta)
X,Y = np.meshgrid(x,y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X-1)**2 - (Y - 1)**2)
Z = (Z1-Z2) * 2
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X,Y,Z)
ax.clabel(CS,inline = 1,fontsize = 10)
ax.set_title("contour map")
```

# Draw 3d surface image

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
fig = plt.figure(figsize =(14,6),dpi = 300)

#specify the 3D graphics to draw, with projection = "3D"
ax = fig.add_subplot(1,2,1,projection = "3d")
ax.plot_surface(X,Y,Z, rstride= 4, cstride =4,lw = 1)
```
